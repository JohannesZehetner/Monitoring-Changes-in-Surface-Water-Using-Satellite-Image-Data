```
!pip install rasterio
import rasterio
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt


raster = rasterio.open('/content/drive/My Drive/Satellite Image/1.jpeg')
r, g, b  = raster.read()
total = np.dstack((r, g, b))
tfImage_rot=tf.image.rot90(total)

plt.imshow(total)
plt.show() 
plt.imshow(tfImage_rot)
plt.show()  

```
